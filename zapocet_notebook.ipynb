```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
import scipy.integrate
import random
import math
import funkce
```

```python
funkce.plot_funkce(funkce.f_1, 0, np.pi, title="Vykres funkce $\sin(x) \exp(\cos(x))$")
funkce.plot_funkce(funkce.f_2, 0, 2 * np.pi, title="Vykres funkce $\sin^{6}(x)$")
funkce.plot_funkce(funkce.f_3, 0, 4, title="Vykres funkce $x^{2}$")
```

Standardni metoda Monte Carlo

```python
print("Odhad integralu f_1 metodou Monte Carlo pro 10000 kapek: ")
print(funkce.monte_carlo(funkce.f_1, 10000, 0, np.pi))
print("Odhad integralu f_2 metodou Monte Carlo pro 10000 kapek: ")
print(funkce.monte_carlo(funkce.f_2, 10000, 0, 2 * np.pi))
print("Odhad integralu f_3 metodou Monte Carlo pro 10000 kapek: ")
print(funkce.monte_carlo(funkce.f_3, 10000, 0, 4 ))

print("Numericka hodnota integralu f_1: ")
result, error = scipy.integrate.quad(funkce.f_1, 0, np.pi)
print(result)
print("Numericka hodnota integralu f_2: ")
result, error = scipy.integrate.quad(funkce.f_2, 0, 2 * np.pi)
print(result)
print("Numericka hodnota integralu f_3: ")
result, error = scipy.integrate.quad(funkce.f_3, 0, 4)
print(result)
```

Vylepsena metoda Monte Carlo

Horn obdelnkov odhad funkce f na intervalu [start, end]
- rozdlen intervalu [start, end] na n stejn velkch podinterval
- nalezen maxim funkce f na kadm z podinterval
- horn odhad funkce f na [start, end] maximy

tedy:
$$
{\displaystyle f(x)=\max f([x_{i-1},x_{i}])}  \ \forall \, x \in [x_{i-1}, x_{i}]
$$ 




```python
funkce.plot_horni_odhad(funkce.f_1, 20, 0, np.pi)
funkce.plot_horni_odhad(funkce.f_2, 20, 0, 2 * np.pi)
funkce.plot_horni_odhad(funkce.f_3, 20, 0, 4)
```

- spoten odhadu integrlu v kadm podintervalu metodou Monte Carlo
- seten dlch vsledk

```python
print("Hodnota integralu f_1 spoctena vylepsenou metodou Monte Carlo: ")
print(funkce.integrace(funkce.f_1, 20, 5000, 0, np.pi))
print("Hodnota integralu f_2 spoctena vylepsenou metodou Monte Carlo: ")
print(funkce.integrace(funkce.f_2, 20, 5000, 0, 2 * np.pi))
print("Hodnota integralu f_3 spoctena vylepsenou metodou Monte Carlo: ")
print(funkce.integrace(funkce.f_3, 20, 5000, 0, 4))

print("Numericka hodnota integralu f_1: ")
result, error = scipy.integrate.quad(funkce.f_1, 0, np.pi)
print(result)
print("Numericka hodnota integralu f_2: ")
result, error = scipy.integrate.quad(funkce.f_2, 0, 2 * np.pi)
print(result)
print("Numericka hodnota integralu f_3: ")
result, error = scipy.integrate.quad(funkce.f_3, 0, 4)
print(result)
```
